```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python


doc = pd.read_excel('benchmark.xlsx', sheet_name='PETR4')
plt.plot(doc['x'], doc['z'], color='red')

doc = pd.read_excel('benchmark.xlsx', sheet_name='BOVA11') 
plt.plot(doc['x'], doc['z'], color='blue')


plt.title('PETR4 X BOVA11', size='20', color='black')
plt.legend(['PETR4','BOVA11'], loc = 'upper left')
plt.axis((0,30,85,115))
plt.xlabel('Setembro', size='15')
plt.ylabel('Variação',size='15')
plt.grid()

```
