```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

doc = pd.read_excel('bench.xlsx', sheet_name='PETR4')
plt.plot(doc['x'], doc['v'], color='red')

doc = pd.read_excel('bench.xlsx', sheet_name='BOVA11') 
plt.plot(doc['x'], doc['v'], color='blue')


plt.title('PETR4 X BOVA11', size='20', color='black')
plt.legend(['PETR4','BOVA11'], loc = 'lower left')
plt.axis((0,30,-0.15,0.15))
plt.xlabel('Setembro', size='15')
plt.ylabel('Variação',size='15')
plt.grid()

```
